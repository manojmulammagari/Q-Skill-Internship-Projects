# Matrix Operations Tool
This tool demonstrates basic and advanced matrix operations using NumPy.

```python
import numpy as np

# Define sample matrices for the operations
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
```

```python
# Basic Arithmetic: Addition and Subtraction
sum_result = np.add(A, B)
diff_result = np.subtract(A, B)

print("Addition (A + B):")
print(sum_result)
print("\nSubtraction (A - B):")
print(diff_result)
```

```python
# Matrix Multiplication (Dot Product)
prod_result = np.dot(A, B)

print("Matrix Multiplication (A * B):")
print(prod_result)
```

```python
# Transpose and Determinant
transpose_A = np.transpose(A)
det_A = np.linalg.det(A)

print("Transpose of A:")
print(transpose_A)
print(f"\nDeterminant of A: {det_A}")
```

```python
# Matrix Inverse
try:
    inv_A = np.linalg.inv(A)
    print("Inverse of A:")
    print(inv_A)
except np.linalg.LinAlgError:
    print("Matrix A is not invertible.")
```

```python
def get_matrix(name):
    print(f"Enter 4 numbers for 2x2 Matrix {name} (space-separated):")
    data = input().split()
    return np.array(data, dtype=float).reshape(2, 2)

print("--- Matrix Tool Menu ---")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Transpose")
print("5. Determinant")

choice = input("Select an operation (1-5): ")

if choice in ['1', '2', '3']:
    m1 = get_matrix("A")
    m2 = get_matrix("B")
    if choice == '1':
        print("\nResult (Addition):\n", m1 + m2)
    elif choice == '2':
        print("\nResult (Subtraction):\n", m1 - m2)
    elif choice == '3':
        print("\nResult (Multiplication):\n", np.dot(m1, m2))

elif choice in ['4', '5']:
    m = get_matrix("A")
    if choice == '4':
        print("\nResult (Transpose):\n", m.T)
    elif choice == '5':
        print("\nResult (Determinant):", np.linalg.det(m))
else:
    print("Invalid choice selected.")
```
